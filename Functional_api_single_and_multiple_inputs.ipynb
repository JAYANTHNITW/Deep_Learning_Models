```python
from keras.models import Model
```

```python
from keras.layers import *
```

### Model with single input and multiple outputs

```python
x = Input(shape=(3,1))
```

```python
hidden1= Dense(128,activation='relu')(x)
hidden2 = Dense(64,activation='relu')(hidden1)

output1 = Dense(1,activation='linear')(hidden2)
output2= Dense(1,activation='sigmoid')(hidden2)
```

```python
model = Model(inputs=x, outputs=[output1,output2])
```

```python
model.summary()
```

```python
from keras.utils import plot_model
plot_model(model,show_shapes=True)
```

### Model with multiple inputs and single input

```python
from keras.layers import *
```

```python
# Let's define the inputs
input1 = Input(shape=(12,))
input2 = Input(shape=(24,))
```

```python
# The first Branch opertes on the first input
x = Dense(32,activation='relu')(input1)
x1 = Dense(16,activation='relu')(x)
x2 = Dense(8,activation='relu')(x1)
x3 = Dense(4,activation='relu')(x2)
```

```python
# The Second branch opreates on the second input
y = Dense(120,activation='relu')(input2)
y1 = Dense(60,activation='relu')(y)
y2 = Dense(30,activation='relu')(y1)
y3 = Dense(15,activation='relu')(y2)
```

```python
# combine the output of the two branches
combined = concatenate([x3,y3])
```

```python
# apply a FC layer and then a regression prediction on the combined outputs

z = Dense(2,activation='relu')(combined)
z1 = Dense(1,activation='linear')(z)
```

```python
model = Model(inputs=[input1,input2],outputs=z1)
```

```python
plot_model(model,show_shapes=True)
```
